```python
import utils
import instance
import numpy as np

X, theta = instance.soare(2, .2)
Y = np.array([(X[0] - X[i])/(theta@(X[0] - X[i])) for i in range(1,len(X))])
a = utils.FW(X, Y, step_size=1, iters=50000,verbose=True, logging_step=100000)
l = a[0]
def alt(best_idx_2, best_idx):
    xstar = X[best_idx]
    z = X[best_idx_2]
    A = sum([np.outer(X[i], X[i])*l[i] for i in range(X.shape[0])])
    A = np.linalg.inv(A)
    gap = (xstar-z)@theta
    var = (xstar-z).T@A@(xstar-z)
    a = gap**2*(xstar-z).T@A@(xstar-z)/var**2
    return a
print(alt(1,0))
print(alt(2,0))
print(max(a[1]*2))
print(a[0])
```

```python
l = a[0]
A = np.array([X[0] - X[1], X[0] - X[2]] )
Al = l[0]*np.outer(X[0],X[0]) + l[1]*np.outer(X[1],X[1]) + l[2]*np.outer(X[2],X[2]) 
for t in range(1,10000):
    p = utils.prob_region(A, theta, np.linalg.inv(Al*t))
    rho = (t+1)/(-np.log(1-p))
    if t%100==0:
        print('probability', p, rho)
```

```python
import ast
arrays = []

with open('/home/lalitj/contextualbandits/bandit-elimination-main/experiments/results_c.txt', 'r') as f:
    content = f.read().split('	')  # Assuming that tuples are separated by 4 spaces
    for item in content:
        #print(item)
        tuple_data = ast.literal_eval(item)
        arrays.append([np.array(list_data) for list_data in tuple_data])

n_iters = len(arrays[0][0])
recs = np.ones([len(arrays), n_iters+1])

for i in range(len(arrays)):
    arr = arrays[i]
    recs_item = np.array(arr[1])
    recs[i,0:(len(recs_item)-1)] = recs_item[1:]

idx_star = 1
reps = len(arrays)
xaxis_g = np.arange(n_iters+1)

m_g = (recs == idx_star).mean(axis=0)
s_g = (recs == idx_star).std(axis=0)/np.sqrt(reps)
```
