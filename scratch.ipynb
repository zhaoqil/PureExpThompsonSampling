```python
import numpy as np

def pairwise_diff(arr):
    """Compute all unique pairwise differences between elements in a numpy array"""
    # Reshape the input array as a column vector
    arr = arr.reshape((-1, 1))
    # Compute the differences between all pairs of elements
    diffs = arr - arr.T
    # Return the unique non-zero differences
    return np.unique(diffs[np.nonzero(diffs)])

# Example usage:
arr = np.array([1, 2, 3])
diffs = pairwise_diff(arr)
print(diffs)
```
