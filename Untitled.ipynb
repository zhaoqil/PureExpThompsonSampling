```python
import numpy as np
from distribution import Gaussian, GenericFunction
from bandit_type import Linear
from instance import soare
from library import *
import library
import utils
import importlib
```

```python

class General(Linear):
    
    def __init__(self, X, Y, gen_star, T, sigma, name):
        super().__init__(X, Y, gen_star, T, sigma, name)
        self.pi = Gaussian(np.zeros(self.d), self.V)
        if type(gen_star) is np.ndarray:
            self.gen_star = GenericFunction(lambda x: x@gen_star, sigma)
        else:
            self.gen_star = gen_star
        self.B = 20
        self.pulled = []
        self.name = name
        self.delta=.001
        
    def run(self, logging_period=1, k=10):
        for t in range(self.T):

            f1 = self.pi.sample()
            idx1 = np.argmax(f1.evaluate(self.X))
            x1 = self.X[idx1]

            a = 0
            idx2 = idx1
            while idx1 == idx2:
                f2s = self.pi.sample(k)  # TODO: sample 10 at a time
                for f2 in f2s:
                    idx2 = np.argmax(f2.evaluate(self.X))
                    if idx1 != idx2:
                        break
                a+=k
                if a > 1/self.delta:
                    break
            if a > 1/self.delta:
                    break
                    
            x2 = self.X[idx2]

            v = []               
            for idx in range(self.n):
                x = self.X[idx]
                expected_diff = 0
                expected_diff_squared = 0
                
                for b1 in range(self.B):
                    gen_b1 = self.pi.sample()
                    y_b1 = gen_b1.pull(x)
                    weight = t # we should change this weighting to simulating t observations from x
                    pi_plus = self.pi.update_posterior(x*weight, y_b1, copy=True)
                    gen_b2 = pi_plus.sample()
                    expected_diff += ( gen_b2.evaluate(x1) - gen_b2.evaluate(x2) ) 
                    expected_diff_squared += ( gen_b2.evaluate(x1) - gen_b2.evaluate(x2) )**2
                v.append( expected_diff_squared/self.B - (expected_diff/self.B)**2 )
                
            min_idx = np.argmin(v)
            self.pulled.append(min_idx)
            x_n = self.X[min_idx]
            y_n = self.gen_star.pull(x_n)
            self.pi.update_posterior(x_n, y_n)
            
            fhat = self.pi.map()
            idx = np.argmax(fhat.evaluate(self.X))
            self.arms_recommended.append(idx)

            if t%logging_period == 0:
                print('general run', self.name, 'iter', t, "/", self.T, end="\r")
```

## Demonstrating posterior updates and sampling with Catboost

```python
import numpy as np
from catboost import CatBoostRegressor
import matplotlib.pyplot as plt
from scipy.stats import truncnorm
import library_concept
from bandit_type import Concept
import distribution

# create truncated normal distribution object
mean = 0.5  # mean of the distribution
std_dev = 0.5  # standard deviation of the distribution
lower_bound = 0  # lower bound of the truncation range
upper_bound = 1  # upper bound of the truncation range
a = (lower_bound - mean) / std_dev
b = (upper_bound - mean) / std_dev
trunc_norm = truncnorm(a, b, loc=mean, scale=std_dev)

# generate samples
f_star = lambda x: -x*np.log(x) - (1-x)*np.log(1-x)
sigma = 0.2
d = 1

pi = library_concept.Kitten(np.empty((0,d)), np.empty((0)), f_star, sigma)

num_samples = 25
X = trunc_norm.rvs(num_samples)
Y = f_star(X) +sigma*np.random.randn(len(X))

pi.update_posterior(X,Y)

results =[]
xaxis = np.linspace(1e-4,1-1e-4,100).reshape(100,1)
for i in range(20):
    f = pi.sample()
    results.append(f.evaluate(xaxis))
mu = np.mean(results, axis=0)
std = np.std(results, axis=0)
plt.figure()
plt.scatter(X,Y)
plt.fill_between(xaxis.reshape(-1), mu-std*2, mu+std*2, alpha=0.3)
plt.plot(xaxis, mu,label='ensemble average')
plt.plot(xaxis, f_star(xaxis),'r',label='truth')
plt.title('sigma = %.2f' % (sigma))
plt.legend()
plt.show()
```
